## Digital Library Class

```python
# noinspection PyTypeChecker
class DigitalLibrary:
    """
    Main GUI controller for the Digital Library application.

    This class initializes the main Tkinter window, builds the primary
    interface layout, and manages the central content container used by
    library-related views.

    Attributes:
        window: Root Tk/ttkbootstrap window instance.
        container: Main frame that holds dynamic UI components.

    Methods:
        build_ui(): Configure the window and create the base widgets.
        destroy_examples(): Remove all widgets from the main container.
    """
    def __init__(self, tkwindow):
        self.window = tkwindow
        self.container = None
        self.menu_map = None

        self.build_menu_map()
        self.build_ui()

    def build_menu_map(self):
        self.menu_map = {
            "Book Management": BookManagement,
            "User Management": UserManagement,
            "Loan": LoanManagement,
            "Statistics and Reports": StatisticsReports,
        }

    def build_ui(self):
        self.window.title("Digital Library")
        self.window.geometry("1680x900")

        control_frame = ttk.Frame(self.window)
        control_frame.pack(fill=tk.X, padx=30, pady=10)

        # Title
        ttk.Label(master=control_frame, text="Digital Library", style="h5.TLabel").pack(
            side=tk.LEFT,
            padx=0,
            fill=tk.X,
            expand=True)

        # Quit button
        ttk.Button(master=control_frame, text="Quit", command=self.window.destroy, width=10, bootstyle="danger").pack(
            side=tk.LEFT,
            padx=5)

        ttk.Separator(master=self.window, orient=tk.HORIZONTAL, style='default.Horizontal.TSeparator').pack(padx=30, fill=tk.X)

        # Menu
        menu_frame = ttk.Frame(self.window)
        menu_frame.pack(fill=tk.X, padx=30, pady=10)

        # Menu buttons
        for menu in self.menu_map:
            ttk.Button(master=menu_frame, text=menu, width=46, command=lambda feature=menu: self.load_feature(feature)).pack(
                side=tk.LEFT,
                padx=5)

        ttk.Separator(master=self.window, orient=tk.HORIZONTAL, style='default.Horizontal.TSeparator').pack(padx=30, fill=tk.X)

        self.container = tk.Frame(self.window)
        self.container.pack(fill=tk.BOTH, padx=35, pady=20, expand=True)

    def load_feature(self, feature):
        frame_class = self.menu_map.get(feature)

        self.destroy_feature()

        if frame_class is not None:
            frame = frame_class(self.container)
            frame.pack(fill=tk.BOTH, expand=True)

    def destroy_feature(self):
        for widget in self.container.winfo_children():
            widget.destroy()



def configure_style(style):
    style.configure("h1.TLabel", font=("Helvetica", 48, "bold"))
    style.configure("h2.TLabel", font=("Helvetica", 36, "bold"))
    style.configure("h3.TLabel", font=("Helvetica", 32, "bold"))
    style.configure("h4.TLabel", font=("Helvetica", 28, "bold"))
    style.configure("h5.TLabel", font=("Helvetica", 24, "bold"))
    style.configure("h6.TLabel", font=("Helvetica", 20, "bold"))
    style.configure("h7.TLabel", font=("Helvetica", 18))
    style.configure("h8.TLabel", font=("Helvetica", 16))
    style.configure("h9.TLabel", font=("Helvetica", 14))
    style.configure("normal.TLabel", font=("Helvetica", 12))
```

## Start Application

```python
ttk.Style.instance = None
root = ttk.Window(themename="darkly")
configure_style(root.style)
DigitalLibrary(root)
root.mainloop()
root.update_idletasks()
del root
```
