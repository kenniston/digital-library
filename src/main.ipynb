## Digital Library Class

```python
# noinspection PyTypeChecker
class DigitalLibrary:
    """
    Main GUI controller for the Digital Library application.

    This class initializes the main Tkinter window, builds the primary
    interface layout, and manages the central content container used by
    library-related views.

    Attributes:
        window: Root Tk/ttkbootstrap window instance.
        container: Main frame that holds dynamic UI components.

    Methods:
        build_ui(): Configure the window and create the base widgets.
        destroy_examples(): Remove all widgets from the main container.
    """
    def __init__(self, tkwindow):
        self.window = tkwindow
        self.container = None
        self.menu_map = None
        self.categories = self._default_categories()
        self.users = self._default_users()
        self.loans = []
        self.book_totals = {}
        self.create_sample_data()

        self.build_menu_map()
        self.build_ui()

    @staticmethod
    def _default_categories() -> dict:
        return {
            "Fiction": [],
            "Science Fiction": [],
            "Fantasy": [],
            "Mystery": [],
            "Romance": [],
            "Technology": [],
            "Business": [],
            "Education": []
        }

    @staticmethod
    def _default_users() -> dict:
        return {
            "Student": [],
            "Professor": [],
            "Visitor": []
        }

    def create_sample_data(self) -> None:
        """Populate default sample books/users in global shared mappings."""
        if sum(len(books) for books in self.categories.values()) == 0:
            self.categories["Fiction"].append(Book("Dune", "Frank Herbert", "9780441172719", 1965, 4))
            self.categories["Science Fiction"].append(Book("Dune Messiah", "Frank Herbert", "9780441172696", 1969, 3))
            self.categories["Science Fiction"].append(Book("Children of Dune", "Frank Herbert", "9780441104024", 1976, 2))
            self.categories["Fiction"].append(Book("The Hobbit", "J.R.R. Tolkien", "9780547928227", 1937, 3))
            self.categories["Mystery"].append(Book("Gone Girl", "Gillian Flynn", "9780307588371", 2012, 2))
            self.categories["Technology"].append(Book("Clean Code", "Robert C. Martin", "9780132350884", 2008, 5))
            self.categories["Technology"].append(Book("Clean Architecture", "Robert C. Martin", "9780134494166", 2017, 4))

        if sum(len(users) for users in self.users.values()) == 0:
            self.users[UserType.PROFESSOR.label].append(User("Ana Pereira", "ana.pereira@gmail.com", "Rua A, 100", str(uuid.uuid4()), UserType.PROFESSOR))
            self.users[UserType.STUDENT.label].append(User("Bruno Lima", "bruno.lima@gmail.com", "Rua B, 220", str(uuid.uuid4()), UserType.STUDENT))
            self.users[UserType.VISITOR.label].append(User("Carla Souza", "carla.souza@gmail.com", "Rua C, 33", str(uuid.uuid4()), UserType.VISITOR))

        if len(self.book_totals) == 0:
            for books in self.categories.values():
                for book in books:
                    isbn = str(getattr(book, "isbn", "")).strip()
                    if isbn:
                        self.book_totals[isbn] = int(getattr(book, "quantity", 0))

        if len(self.loans) == 0:
            today = date.today()
            sample_loans = [
                {
                    "user_name": "Ana Pereira",
                    "user_email": "ana.pereira@gmail.com",
                    "book_title": "Dune",
                    "isbn": "9780441172719",
                    "loan_date": (today - timedelta(days=5)).isoformat(),
                    "due_date": (today + timedelta(days=25)).isoformat(),
                    "returned": False
                },
                {
                    "user_name": "Bruno Lima",
                    "user_email": "bruno.lima@gmail.com",
                    "book_title": "Clean Code",
                    "isbn": "9780132350884",
                    "loan_date": (today - timedelta(days=10)).isoformat(),
                    "due_date": (today + timedelta(days=20)).isoformat(),
                    "returned": False
                },
                {
                    "user_name": "Carla Souza",
                    "user_email": "carla.souza@gmail.com",
                    "book_title": "The Hobbit",
                    "isbn": "9780547928227",
                    "loan_date": (today - timedelta(days=40)).isoformat(),
                    "due_date": (today - timedelta(days=10)).isoformat(),
                    "returned": True
                }
            ]
            self.loans.extend(sample_loans)

            open_loans_by_isbn = {}
            for loan in self.loans:
                if bool(loan.get("returned", False)):
                    continue
                isbn = str(loan.get("isbn", "")).strip()
                if not isbn:
                    continue
                open_loans_by_isbn[isbn] = open_loans_by_isbn.get(isbn, 0) + 1

            for books in self.categories.values():
                for book in books:
                    isbn = str(getattr(book, "isbn", "")).strip()
                    if not isbn:
                        continue
                    total = int(self.book_totals.get(isbn, int(getattr(book, "quantity", 0))))
                    borrowed = int(open_loans_by_isbn.get(isbn, 0))
                    book.quantity = max(0, total - borrowed)

    def build_menu_map(self):
        self.menu_map = {
            "📚 Book Management": lambda parent: BookManagement(parent, shared_categories=self.categories),
            "👥 User Management": lambda parent: UserManagement(parent, shared_users=self.users),
            "🏷️ Loan": lambda parent: LoanManagement(
                parent,
                shared_categories=self.categories,
                shared_users=self.users,
                shared_loans=self.loans,
                shared_book_totals=self.book_totals
            ),
            "📈 Statistics and Reports": lambda parent: StatisticsReports(
                parent,
                shared_categories=self.categories,
                shared_users=self.users,
                shared_loans=self.loans
            ),
        }

    def build_ui(self):
        self.window.title("Digital Library")
        self.window.geometry("1680x900")

        control_frame = ttk.Frame(self.window)
        control_frame.pack(fill=tk.X, padx=30, pady=10)

        # Title
        ttk.Label(master=control_frame, text="Digital Library", style="h5.TLabel").pack(
            side=tk.LEFT,
            padx=0,
            fill=tk.X,
            expand=True)

        # Quit button
        ttk.Button(master=control_frame, text="Quit", command=self.window.destroy, width=10, bootstyle="danger").pack(
            side=tk.LEFT,
            padx=5)

        ttk.Separator(master=self.window, orient=tk.HORIZONTAL, style='default.Horizontal.TSeparator').pack(padx=30, fill=tk.X)

        # Menu
        menu_frame = ttk.Frame(self.window)
        menu_frame.pack(fill=tk.X, padx=30, pady=10)

        # Menu buttons
        for menu in self.menu_map:
            ttk.Button(master=menu_frame, text=menu, width=46, command=lambda feature=menu: self.load_feature(feature)).pack(
                side=tk.LEFT,
                padx=5)

        ttk.Separator(master=self.window, orient=tk.HORIZONTAL, style='default.Horizontal.TSeparator').pack(padx=30, fill=tk.X)

        self.container = tk.Frame(self.window)
        self.container.pack(fill=tk.BOTH, padx=35, pady=20, expand=True)

    def load_feature(self, feature):
        frame_factory = self.menu_map.get(feature)

        self.destroy_feature()

        if frame_factory is not None:
            frame = frame_factory(self.container)
            frame.pack(fill=tk.BOTH, expand=True)

    def destroy_feature(self):
        for widget in self.container.winfo_children():
            widget.destroy()



def configure_style(style):
    style.configure("h1.TLabel", font=("Helvetica", 48, "bold"))
    style.configure("h2.TLabel", font=("Helvetica", 36, "bold"))
    style.configure("h3.TLabel", font=("Helvetica", 32, "bold"))
    style.configure("h4.TLabel", font=("Helvetica", 28, "bold"))
    style.configure("h5.TLabel", font=("Helvetica", 24, "bold"))
    style.configure("h6.TLabel", font=("Helvetica", 20, "bold"))
    style.configure("h7.TLabel", font=("Helvetica", 18))
    style.configure("h8.TLabel", font=("Helvetica", 16))
    style.configure("h9.TLabel", font=("Helvetica", 14))
    style.configure("normal.TLabel", font=("Helvetica", 12))
```

## Start Application

```python
ttk.Style.instance = None
root = ttk.Window(themename="darkly")
configure_style(root.style)
DigitalLibrary(root)
root.mainloop()
root.update_idletasks()
del root
```
