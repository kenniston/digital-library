# Base Class for Digital Library Features

## Import Libraries

```python
import pandas as pd
import tkinter as tk
import ttkbootstrap as ttk
from tkinter import Frame
import pandastable as pt
```

## Base Frame Class

```python
# noinspection PyTypeChecker
class BaseFrame(Frame):
    def __init__(self, master=None):
        super().__init__(master)
        self._master = master
        self.data = None
        self.table = None
        self.left_panel = None
        self.right_panel = None
        self.left_panel_container = None

        self._build_ui()

    def _build_ui(self):
        # Content Panel
        canvas = tk.Canvas(self)
        scrollbar = tk.Scrollbar(self, orient="vertical", bg="#2c2c2c",
                                 troughcolor="#252525",
                                 activebackground="#353535",
                                 bd=0,
                                 highlightthickness=0,
                                 command=canvas.yview)
        canvas.configure(yscrollcommand=scrollbar.set)

        canvas.pack(side=tk.LEFT, fill=tk.BOTH, expand=True)
        scrollbar.pack(side=tk.LEFT, fill=tk.Y, pady=20)

        self.left_panel = Frame(master=canvas)
        window = canvas.create_window((0, 0), window=self.left_panel, anchor="nw")

        def on_configure(event):
            canvas.configure(scrollregion=canvas.bbox("all"))

        self.left_panel.bind("<Configure>", on_configure)

        def resize_frame(event):
            canvas.itemconfig(window, width=event.width)

        canvas.bind("<Configure>", resize_frame)

        # self.left_panel.configure(background="yellow")

        # Main Section Separator
        main_separator = tk.Frame(master=self, width=1.2)
        main_separator.pack(side=tk.LEFT, pady=20, padx=10, fill=tk.Y)
        main_separator.configure(background="#444444")

        # Action Pane
        self.right_panel = Frame(master=self, width=250)
        # self.right_panel.configure(background="blue")
        self.right_panel.pack(side=tk.RIGHT, fill=tk.Y)
        self.right_panel.pack_propagate(False)

        # Build Action Pane
        self._internal_build_right_pane()

        # Build Left Panel
        self._internal_build_left_panel()

    def _internal_build_right_pane(self):
        # Title
        ttk.Label(master=self.right_panel, text="Actions", style="h7.TLabel").pack(pady=10)

        # Title Separator
        title_separator = tk.Frame(master=self.right_panel, height=1)
        # title_separator.pack(pady=0, fill=tk.X)
        title_separator.configure(background="#444444")

        # Build Right Panel
        self.build_right_panel()

    def _internal_build_left_panel(self):
        # Title
        ttk.Label(master=self.left_panel, text=self.get_title(), style="h7.TLabel").pack(pady=10)

        # Title Separator
        title_separator = tk.Frame(master=self.left_panel, height=1, width=800)
        # title_separator.pack(pady=0)
        title_separator.configure(background="#444444")

        self.left_panel_container = tk.Frame(master=self.left_panel)
        self.left_panel_container.pack(fill=tk.BOTH, expand=True)
        self.left_panel_container.pack_propagate(False)
        for i in range(3):
            self.left_panel_container.grid_rowconfigure(i, weight=1)
            self.left_panel_container.grid_columnconfigure(i, weight=1)

        # Build Left Panel
        self.build_left_panel()

    def create_data_table(self, grid, row, column, rowspan, colspan, showtoolbar=False, showstatusbar=True, sticky=tk.NSEW,
                          padx=1, pady=1, title="Data Table", editable=False, enable_menus=False, height=250, width=300):
        # Create the cell panel
        master = Frame(master=grid)
        master.grid(row=row, column=column, rowspan=rowspan, columnspan=colspan, sticky=sticky, padx=padx, pady=pady)

        # Title
        ttk.Label(master=master, text=title, style="h9.TLabel").pack(fill=tk.X)

        # Title Separator
        title_separator = tk.Frame(master=master, height=1, width=300)
        title_separator.pack(pady=(0, 20), anchor=tk.W)
        title_separator.configure(background="#444444")

        table_frame = Frame(master=master, padx=1, pady=1)

        # table_frame.configure(bg="red")
        table_frame.pack(fill=tk.BOTH, expand=True)

        self.table = pt.Table(table_frame, dataframe=self.data, showtoolbar=showtoolbar, showstatusbar=showstatusbar,
                              editable=editable, enable_menus=enable_menus, height=height, width=width)

        options = {
            'align': 'w',
            'cellbackgr': '#222222',
            'cellwidth': 200,
            'floatprecision': 0,
            'thousandseparator': '',
            'grid_color': '#949494',
            'linewidth': 0,
            'rowheight': 28,
            'horizlines': True,
            'vertlines': True,
            'showindex': False,

            'multipleselectioncolor': '#345475',
            'boxoutlinecolor': '#222222',

            'textcolor': 'white',
            'font': 'Arial',
            'fontsize': 10,
            'fontstyle': '',

            'rowheaderbgcolor': '#345475',
            'rowheaderfgcolor': 'white',
            'rowselectedcolor': '#345475',

            'colheaderbgcolor': '#373737',
            'colheadertextcolor': 'white',
            'colselectedcolor': '#345475',
        }
        pt.config.apply_options(options, self.table)
        self.table.show()

    def set_data(self, data: pd.DataFrame):
        self.data = data
        if len(self.data) == 0:
            empty_row = pd.DataFrame([[None] * len(self.data.columns)], columns=self.data.columns)
            self.data = pd.concat([self.data, empty_row], ignore_index=True)

        if self.table is not None:
            self.table.model.df = self.data
            self.table.redraw()

    def build_left_panel(self):
        pass

    def build_right_panel(self):
        pass

    def load_data(self):
        pass

    def get_title(self) -> str:
        pass
```
